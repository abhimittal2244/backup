```python
def GCD(x, y):
    lst = []
    if x<y:
        small=x
    else:
        small=y
    for i in range(1, small+1):
        if x%i == 0 and y%i == 0:
            lst.append(i)
    return lst.pop()

x, y = eval(input("enter the number: "))
print(GCD(x, y))            
```

```python
def calculate_total_average(marks):
    total = sum(marks)
    average = total / len(marks)
    return total, average

# Dictionary containing names of students and marks in three subjects
students = {
    'Alice': [80, 75, 90],
    'Bob': [85, 90, 88],
    'Charlie': [78, 92, 87],
    'David': [92, 88, 95],
}

# Replace marks with total and average
for student, marks in students.items():
    total, average = calculate_total_average(marks)
    students[student] = {'Total': total, 'Average': average}

# Find the topper of the class
topper = max(students, key=lambda student: students[student]['Total'])

# Display the updated dictionary and the topper
print("Updated Dictionary:")
for student, marks in students.items():
    print(f"{student}: {marks}")

print("\nTopper of the class:", topper)
```

```python
class Complex:
    def __init__(self,real,imag):
        self.real=real
        self.imag=imag
    def add(self,obj):
        r = self.real+obj.real
        i = self.imag+obj.imag
        z = Complex(r, i)
        return z
    def sub(self,obj):
        r = self.real-obj.real
        i = self.imag-obj.imag
        z = Complex(r, i)
        return z
    def mul(self,obj):
        r = self.real*self.imag-obj.real*obj.imag
        i = self.real*obj.imag+obj.real*self.imag
        z = Complex(r, i)
        return z
c=Complex(1,2)
d=Complex(2,3)
e = c.add(d)
f = c.sub(d)
g = c.mul(d)

print(e.real, e.imag)
        
```

```python
class complex:
    def __init__(self,real,imag):
        self.real=real
        self.imag=imag
    def add(self,obj):
        print(self.real+obj.real),(self.imag+obj.imag)
    def sub(self,obj):
        print(self.real-obj.real),(self.imag-obj.imag)
    def mul(self,obj):
        print(self.real*self.imag-obj.real*obj.imag) ,(self.real*obj.imag+obj.real*self.imag)
    def conjugate(self):
        print ('{}-{}'.format(self.real, self.imag))
c=complex(1,2)
c1=complex(3,4)
c.add(c1)
c.sub(c1)
c.mul(c1)
c.conjugate()
    
```
